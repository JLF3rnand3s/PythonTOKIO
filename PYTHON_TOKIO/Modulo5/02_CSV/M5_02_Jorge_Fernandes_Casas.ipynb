**3) Continuemos con Excel, centrandonos en la parte de la librería openpyxl.styles (la cual proporciona estilo a nuestras hojas de Excel)**
* Poner las cabeceras que acabamos de añadir en negrita (bold)
* Guarda los cambios y comprueba en el excel si se han efectuado

Pista: Aunque hay varias formas de hacerlo, lo más común es definir el estilo que deseas (negrita) y luego aplicárselo a todas las celdas que quieras (iterar por la primera fila de nuestra hoja)

Doc: https://openpyxl.readthedocs.io/en/stable/styles.html

````python
# Completa el ejercicio aquí
from openpyxl.styles import Font 
import openpyxl
# Ruta al archivo .xlsx
ruta_archivo = '02_Excel_data.xlsx'

# Cargar el archivo
archivo = openpyxl.load_workbook(ruta_archivo)
#sacamos el nombre de la hoja de las Olimpiadas
hoja = archivo['Olimpiadas']
# Añadir cabeceras
cabeceras = ["Pais", "Oros", "Platas", "Bronces"]
#sacamos el valor de la negrita de la libreria de openpyxl
negrita = Font(bold=True)

# Iterar sobre las celdas de la primera fila y aplicar formato de negrita0
#iteramos en columna-valor para toda la columna numero 1, el start = 1 es por que empieza en la celda nº1 he de suponer
for columna, valor in enumerate(cabeceras, start=1):
    #creas un valor llamado celda en el cual haras referencia a la celda (fila-columna en la cual empieza tu cabecera)
    celda = hoja.cell(row=1, column=columna)
    #igualamos el valor de la celda que la sacamos en el for
    celda.value = valor
    #a las celdas le añadimos la negrita
    celda.font = negrita
#Creamos una lista
datos_olimpicos = [
    ["USA", 46, 12, 5],
    ["China", 38, 20, 7],
    ["UK", 29, 7, 7],
    ["Russia", 22, 10, 9],
    ["South Korea", 13, 3, 2],
    ["Germany", 11, 7, 4]
]
#iteramos para ingresar todos los datos que tengamos en la lista
for fila in datos_olimpicos:
    hoja.append(fila)

# Listar el contenido por filas y columnas
for fila in hoja.iter_rows(values_only=True):
    print(fila)
    
archivo.save("02_Excel_data.xlsx")
````

**4) Continuamos con Excel. Crea una nueva columna que sea el sumatorio de todas las medallas conseguidas por cada uno de los paises**

````python
from openpyxl.styles import Font 

#añadir el titulo de olimpiadas
hoja_olimpiadas = archivo['Olimpiadas']
hoja_olimpiadas['E1'] = "Total"
hoja_olimpiadas['E1'].font = Font(bold=True)

#realizar las sumas

'''inicializamos la variable i, la cual empezará en 2 ya que pondremos los resultados la fila "E". A su vez 
colocaremos el rango que queremos que luego recorrá el rango en el ciclo "For" (es decir, que se seleccionará las celdas desde la B2 hasta las D7.
No se seleccionaran las filas "A")'''
i = 2
rango = hoja_olimpiadas["B2:D7"] # Especifico el rango en el que quiero que se sumen los valores de las celdas por cada fila
'''realizamos una iteracion (ciclo for) en el cual sumará desde las celdas B hasta la D y como antes se dijo se colocará en las celdas E''' 
for fila in rango:
    hoja_olimpiadas[f'E{i}'] = f'=SUM(B{i}:D{i})' # El resultado como antes se dijo, se guardará en la celda especificada
    i += 1

archivo.save("02_Excel_data.xlsx")
````

**5) Continuamos con Excel. Crear un gráfico de barras (openpyxl.chart.BarChart()) lo más parecido al que se muestra en la imagen:**



````python
import openpyxl
from openpyxl import Workbook
from openpyxl.chart import BarChart, Reference

# Completa el ejercicio aquí
# Ruta al archivo .xlsx
ruta_archivo = '02_Excel_data.xlsx'
# Cargar el archivo
archivo = openpyxl.load_workbook(ruta_archivo)
#sacamos el nombre de la hoja de las Olimpiadas
hoja = archivo['Olimpiadas']

datos = Reference(hoja, min_col=5, min_row=2, max_col=5, max_row=7)
nombres = Reference(hoja, min_col=1,min_row=2,max_row=7)
categs = BarChart()
categs.add_data(datos, titles_from_data=False)
hoja.add_chart(categs, "A10")
categs.type = "col"
categs.style = 10
categs.varyColors = True
categs.legend = None
categs.title = "Medallas Olimpicas"
categs.y_axis.majorGridlines = None
categs.set_categories(nombres)

archivo.save("02_Excel_data.xlsx")
````

**6) Observa los datos del csv y realiza las siguientes tareas:**

* Tenemos 9 columnas, las 8 primeras contienen datos numéricos con los cuales podemos trabajar, crea una lista para cada una de estas columnas (mpg, cylinders, etc.). No es necesario crear lista para name
* Recorre los datos del csv adecuadamente y almacena los datos de cada columna en cada una de las listas que has creado anteriormente.
* Comprobar que se haya guardado en las listas la información correspondiente (con prints)

````python
# Completa el ejercicio aquí
import csv
#Creamos las listas
mpg = []
cylinders = []
displacement = []
horsepower = []
weight = []
acceleration = []
year = []
origin = []
name = []

#abrimos el csv
archivo_csv = "02_CSV_data.csv"
with open(archivo_csv, "r") as archivoPlano:
    reader = csv.reader(archivoPlano)
    next(reader)
    for row in reader:
            
        #TODO = Crear lista para cada una de las columnas a excepcion de "name".
        mpg.append(float(row[0]))  
        cylinders.append(int(row[1]))  
        displacement.append(float(row[2]))  
        horsepower.append(int(row[3]))  
        weight.append(int(row[4]))  
        acceleration.append(float(row[5]))  
        year.append(int(row[6]))  
        origin.append(int(row[7]))  
        
        #Hacemos el de name tambien ya que lo necesitaremos para el proximo ejercicio
        name.append(str(row[8]))

#TODO = hacer prints para confirmar que se hayan guardado los datos
print("mpg:",mpg)
print("cylinders:",cylinders)
print("displacement:",displacement)
print("horsepower:",horsepower)
print("weight:",weight)
print("acceleration:",acceleration)
print("year:",year)
print("origin:",origin)

print("name:", name)

del reader           # Borrar objetos
archivoPlano.close()  # Cerrar archivo
del archivoPlano      # Borrar objeto
````
