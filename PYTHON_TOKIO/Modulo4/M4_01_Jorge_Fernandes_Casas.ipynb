**1) Ejercicio de herencia múltiple y polimorfismo:**
* Crea la clase Vehiculos con un constructor que incluya marca y modelo
* Esta clase Vehiculos también tendrá que incluir un método que se llame repostar() el cual imprima por pantalla "Este vehiculo tiene que repostar gasolina"
* Crea la clase VElectricos con un constructor que incluya marca, modelo y autonomia
* Esta clase VElectricos también tendrá que incluir un método que se llame repostar() el cual imprima por pantalla "Este vehiculo tiene que repostar electricidad"
* Crea la clase BicicletaElectrica que herede de Vehiculos y de VElectricos, pero dando prioridad a VElectricos (ya que es un vehículo más electrico que normal)
* Crea la clase Quad que herede de Vehiculos y de VElectricos, pero dando prioridad a Vehiculos (ya que sólo usa la electricidad de modo puntual)
* Crear un objeto de BicicletaElectrica y otro de Quad
* Emplear las técnicas de polimorfismo aprendidas para conseguir que al peguntarle a los objetos que acabamos de crear por su repostaje, ambos nos respondan adecuadamente



```python
# Crea la clase Vehiculos con un constructor que incluya marca y modelo
class Vehiculo():
    
    def __init__(self, marca, modelo):
        self.marca = marca
        self.modelo = modelo
        print("Vehiculo creado!")
    
    def __str__(self):
        return f"Marca:{self.marca} Modelo: {self.modelo}"
# Esta clase Vehiculos también tendrá que incluir un método que se llame repostar() el cual imprima por pantalla
# "Este vehiculo tiene que repostar gasolina"
    def repostar(self):
        print("Este vehiculo tiene que repostar gasolina")
            
# Crea la clase VElectricos con un constructor que incluya marca, modelo y autonomia
class VElectricos():
    
    def __init__(self, marca, modelo, autonomia):
        self.marca = marca
        self.modelo = modelo
        self.autonomia = autonomia #combustible (electricidad) ejemplo 100
        print("Vehiculo Electrico ha sido creado")
    
    def __str__(self):
        return f"Vehiculo Electrico con marca: {self.marca}/n modelo: {self.modelo}/n autonomia: {self.autonomia}"

# Esta clase VElectricos también tendrá que incluir un método que se llame repostar() 
# el cual imprima por pantalla "Este vehiculo tiene que repostar electricidad"
    def repostar(self):
        print("Este vehiculo tiene que repostar electricidad")
            
# Crea la clase BicicletaElectrica que herede de Vehiculos y de VElectricos,
# pero dando prioridad a VElectricos (ya que es un vehículo más electrico que normal)
class BicicletaElectrica(VElectricos, Vehiculo):
    #init con los atributos de la clase Vehiculo Electrico
    def __init__(self, marca, modelo, autonomia):
        super().__init__(marca, modelo, autonomia)
        print("Vehiculo tipo Bicicleta Electrica creado")

    
    def __str__(self):
        return f"Bicicleta Electrica con los siguientes atributos: {self.marca}, {self.modelo}, {self.autonomia}"
# Crea la clase Quad que herede de Vehiculos y de VElectricos, pero dando prioridad a Vehiculos (ya que sólo usa la electricidad de modo puntual)
class Quad(Vehiculo, VElectricos):
    def __init__(self, marca, modelo, autonomia):
        super().__init__(marca, modelo)
        #hay que hacer esto si no, no puede pillar el atributo "autonomia especifico de Quad"
        VElectricos.__init__(self, marca, modelo, autonomia)
        print("Vehiculo tipo Quad creado")
        
    def __str__(self):
        return f"Quad con los siguientes atributos: {self.marca}, {self.modelo}"

# Crear un objeto de BicicletaElectrica y otro de Quad
bici = BicicletaElectrica(marca="Ford", modelo="Paraguayas", autonomia=50)
quad = Quad(marca="Ford", modelo="dodge", autonomia=100)
# Emplear las técnicas de polimorfismo aprendidas para conseguir 
# que al peguntarle a los objetos que acabamos de crear por su repostaje, ambos nos respondan adecuadamente
bici.repostar()
quad.repostar()
```

**3) Escoge la temática que prefieras y define una estructura de clases y subclases donde practiques la herencia en al menos dos niveles (herencia multi nivel). Crea una estructura de mínimo 5 clases, con sus respectivos constructores y métodos. Crea algunos objetos y comprueba que la herencia este funcionando. 
Si puedes, incluye alguna técnica de polimorfismo.**

```python
# Completa el ejercicio aquí

class Vehiculo:
    def __init__(self, marca, modelo, año):
        self.marca = marca
        self.modelo = modelo
        self.año = año

    def acelerar(self):
        print("El vehículo está acelerando.")

    def frenar(self):
        print("El vehículo está frenando.")


class Automovil(Vehiculo):
    def __init__(self, marca, modelo, año, puertas):
        super().__init__(marca, modelo, año)
        self.puertas = puertas

    def abrir_puertas(self):
        print("Las puertas del automóvil se abrieron.")

    def cerrar_puertas(self):
        print("Las puertas del automóvil se cerraron.")


class Motocicleta(Vehiculo):
    def __init__(self, marca, modelo, año, cilindrada):
        super().__init__(marca, modelo, año)
        self.cilindrada = cilindrada

    def realizar_acrobacia(self):
        print("La motocicleta está realizando una acrobacia.")


class Bicicleta(Vehiculo):
    def __init__(self, marca, modelo, año, tipo):
        super().__init__(marca, modelo, año)
        self.tipo = tipo

    def pedalear(self):
        print("La bicicleta está siendo pedaleada.")


class Camion(Automovil):
    def __init__(self, marca, modelo, año, puertas, carga_maxima):
        super().__init__(marca, modelo, año, puertas)
        self.carga_maxima = carga_maxima

    def transportar_carga(self):
        print("El camión está transportando carga.")


# Creación de objetos y uso de herencia

auto = Automovil("Ford", "Mustang", 2022, 2)
auto.acelerar()
auto.abrir_puertas()

moto = Motocicleta("Honda", "CBR500R", 2021, 500)
moto.frenar()
moto.realizar_acrobacia()

bike = Bicicleta("Trek", "X-Caliber", 2020, "Montaña")
bike.pedalear()

camion = Camion("Volvo", "VNL", 2019, 2, 20000)
camion.cerrar_puertas()
camion.transportar_carga()
```
